```python
#Converting categorcal features into numerical features 
#1st way   +  df['Gender']=pd.get_dummies(df['Gender'],drop_first=1)
# 2nd way-->using map()
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
#Same using map() covert categorical values into numerical values  
df['City_Category'].unique()
df['City_Category']=df['City_Category'].map({'A':1,'B':2,'C':3})
```

```python
df['Age'].unique()
```

```python
#Converting Age categorical features into numeric values 
# we are using map() since using domain knowledge one can definately tell among all group some age group have max rate of shopping and some age group is very lss so we can assign rank to each group o that model can easily grasp the froup
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```
